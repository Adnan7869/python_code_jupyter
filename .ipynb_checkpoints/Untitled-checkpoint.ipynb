```python
numbers = input("Enter a list").split()
frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

for num,freq in frequency.items():
    print(num,"occurs",freq,"times(s)")
```
